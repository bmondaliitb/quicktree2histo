```python
import awkward as ak

a = ak.Array([[1, 2], [3]])              # two events: event0 has 2 items, event1 has 1
b = ak.Array([["x", "y"], ["z"]])        # same jagged shape
z = ak.zip({"num": a, "label": b})
```

```python
print(a)
print(b)
print(z)
```

```python
#print(z.to_list())
# -> [[{'num': 1, 'label': 'x'}, {'num': 2, 'label': 'y'}],
#     [{'num': 3, 'label': 'z'}]]

flat = ak.flatten(z)   # 1D array: one record per item across all events
print(flat)
print(flat.to_list())
# -> [{'num': 1, 'label': 'x'}, {'num': 2, 'label': 'y'}, {'num': 3, 'label': 'z'}]
```
